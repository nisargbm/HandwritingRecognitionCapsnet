# Copy Files

```
import itertools
import glob
import os
import random
import re
import shutil

isihwd_path = "/content/gdrive/My Drive/Nisarg Colab/Datasets/ISIHWD/"
output_dir = "/content/gdrive/My Drive/Nisarg Colab/Datasets/ISIHWD_WRITER/"
set_num = ["set1", "set2", "set3", "set4", "set5"]
# for set_value in set_num:
set_value = set_num[4]
list_of_words = glob.glob(isihwd_path + set_value + "/word_link/" + "*.bmp" )
print (len(list_of_words))
i = 0
for word in list_of_words:
    i += 1
    if i % 50 == 0:
      print ("{} files completed".format(i))
    
    filename = os.path.basename(word)

    match_writer = re.search("writer(\d+)", filename)
    writer_number = match_writer.group(1)
    match_digit = re.search( set_value + "_(\d+)",filename)
    digit_value = match_digit.group(1)
#         print (filename)
#         print (digit_value, filename)
    new_dir = output_dir + "writer_" + writer_number + "/" + "w_" + digit_value + "/"
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
    shutil.copy(word, new_dir + filename)
#     print (word, new_dir + filename)
```
